```python
#Q3 Answer
from sklearn.metrics import mutual_info_score
```

```python
mutual_info_score(df_train.above_average, df_train.neighbourhood_group)
```

```python
def mutual_info_score_price(series):
    return mutual_info_score(series, df_train.above_average)
```

```python
categorical = ['neighbourhood_group', 'room_type']
mi = df_train[categorical].apply(mutual_info_score_price)
mi.sort_values(ascending = False)
```

```python
#Q4 Answer

from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, random_state=42, max_iter = 1000)
model.fit(X_train, y_train)
```

```python
model_val = LogisticRegression(solver='liblinear', C=1.0, random_state=42, max_iter = 1000)
model_val.fit(X_val, y_val)
```

```python
model.coef_[0].round(3)
```

```python
y_pred = model_val.predict(X_val)
y_pred
```

```python
pred_price = (y_pred >= 152)

pred_price
```

```python
(y_val == y_pred).mean()
```

```python
above_average_pred = [ ]

for price in y_pred:
    if price >= 152:
        ab_ave_pred = 1
    else: ab_ave_pred = 0
    above_average_pred.append(ab_ave_pred)
        
above_average_pred


above_average_val = [ ]

for price in y_val:
    if price >= 152:
        ab_ave_val = 1
    else: ab_ave_val = 0
    above_average_val.append(ab_ave_val)
        
above_average_val


df_pred = pd.DataFrame()
df_pred['prediction'] = y_pred
df_pred['actual'] = y_val
df_pred['above_average_pred'] = above_average_pred
df_pred['above_average_val'] = above_average_val

df_pred['correct'] = df_pred.above_average_pred == df_pred.above_average_val

df_pred.correct.mean().round(2)
```
